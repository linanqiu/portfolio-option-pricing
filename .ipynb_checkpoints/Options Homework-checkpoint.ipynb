```python
import pandas
import math

import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
class Call:
    def __init__(self, strike, direction):
        self.strike = strike
        self.direction = direction
        self.sec = 'Call %s @%s' % (self.direction, self.strike)
    
    def value_at(self, price):
        value = max(price - self.strike, 0)
        if (self.direction == 'buy'):
            return value
        else:
            return -value

class Put:
    def __init__(self, strike, direction):
        self.strike = strike
        self.direction = direction
        self.sec = 'Put %s @%s' % (self.direction, self.strike)
            
    def value_at(self, price):
        value = max(self.strike - price, 0)
        if (self.direction == 'buy'):
            return value
        else:
            return -value
```
